    1. Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

```python
import numpy as np
import matplotlib.pyplot as plt

# Inisialisasi dataset
mostGoals = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
    
print(f"Jumlah Goal: {mostGoals}")

# Mean
mean_Goal = sum(mostGoals) / len(mostGoals)
print("Mean Goal:", mean_Goal)

# Range
range_Goal = max(mostGoals) - min(mostGoals)
print("Range Goal:", range_Goal)

# Variance and Standard Deviation
variance_Goal = sum((x - mean_Goal) ** 2 for x in mostGoals) / len(mostGoals)
std_deviation_Goal = variance_Goal ** 0.5
print("Variance Goal:", variance_Goal)
print("Standard Deviation Goal:", std_deviation_Goal)

plt.hist(mostGoals, bins=20, edgecolor='purple', alpha=0.5, color='black')

# Define the horizontal line coordinates
y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point

# Add a horizontal line
plt.axhline(y=y, xmin=x1, xmax=x2, color='green', linestyle='dashed', linewidth=1, label=f'Range = {range_Goal}')

# Add vertical line
plt.axvline(x=min(mostGoals), color='brown', linestyle='dashed', linewidth=1, label=f'Min Data: {min(mostGoals):.2f}')
plt.axvline(x=max(mostGoals), color='brown', linestyle='dashed', linewidth=1, label=f'Max Data: {max(mostGoals):.2f}')
plt.axvline(x=mean_Goal, color='black', linestyle='dashed', linewidth=1, label=f'Mean = {mean_Goal}')
plt.axvline(x=mean_Goal - std_deviation_Goal, color='blue', linestyle='dashed', linewidth=1, label=f'{mean_Goal-std_deviation_Goal:.2f}')
plt.axvline(x=mean_Goal + std_deviation_Goal, color='blue', linestyle='dashed', linewidth=1, label=f'{mean_Goal+std_deviation_Goal:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')
plt.title(f'Std Deviation = {std_deviation_Goal:.2f}')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
