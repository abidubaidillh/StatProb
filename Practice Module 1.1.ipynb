    1. Diketahui tinggi badan siswa kelas A adalah 100,5; 99,7; 101; 101,3; 99,9; 98,9; 165; 102,2; 98,7; 102,9; 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus? 

```python
import statistics
import matplotlib.pyplot as plt

# Inisialisasi Dataset
data_tinggi_badan_kelas_A = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Fungsi untuk Menghitung Nilai Mean, Median, dan Modus
def nilai_mean(data):
    return sum(data) / len(data)

def nilai_median(data):
    data.sort()
    n = len(data)
    if n % 2 == 0:
        median = (data[n//2 - 1] + data[n//2]) / 2
    else:
        median = data[n//2]
    return median

def nilai_modus(data):
    modus = max(set(data), key = data.count)
    return modus

# Menghitung Nilai Mean, Median, dan Modus
mean_manual = nilai_mean(data_tinggi_badan_kelas_A)

median_manual = nilai_median(data_tinggi_badan_kelas_A)

modus_manual = nilai_modus(data_tinggi_badan_kelas_A)

mean_stats = statistics.mean(data_tinggi_badan_kelas_A)

median_stats = statistics.median(data_tinggi_badan_kelas_A)

modus_stats = statistics.mode(data_tinggi_badan_kelas_A)

# Menampilkan Hasil
print(f"Mean Statistics: {mean_stats}")
print(f"Mean Manual: {mean_manual}")
print(f"Modus Statistics: {modus_stats}")
print(f"Modus Manual: {modus_manual}")
print(f"Median Statistics: {median_stats}")
print(f"Median Manual: {median_manual}")

# Visualisasi Distribusi Data
plt.hist(data_tinggi_badan_kelas_A, bins=15, color='brown', edgecolor='blue')
plt.axvline(mean_stats, color='yellow', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_stats, color='black', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(modus_stats, color='red', linestyle='dashed', linewidth=2, label='Modus')
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```
